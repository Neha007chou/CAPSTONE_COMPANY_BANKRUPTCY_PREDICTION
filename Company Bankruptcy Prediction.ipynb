```python
X = df.drop('Bankrupt', axis = 1)
y = df['Bankrupt']


# X = df.drop('Bankrupt', axis=1): This creates a DataFrame X containing all
#     the columns of df except for the 'Bankrupt' column. 
#     The axis=1 argument specifies that you're dropping a column (as opposed to dropping a row).
# y = df['Bankrupt']: This creates a Series y containing only the 'Bankrupt' column of df.
# This separation is a common step in preparing data for machine learning tasks, where X 
# represents the features or inputs to the model, 
# and y represents the target variable or output that the model aims to predict.
```

```python
from sklearn.decomposition import PCA 

#defining PCA 
pca = PCA().fit(X)

plt.figure(figsize = (10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative sum - Explained Variance')

#fitting PCA on our X
pca = PCA(n_components=0.99).fit(X)
print(f'Best value of n_component for 99% variance : {pca.n_components_}')

pca = PCA(n_components=0.95).fit(X)
print(f'Best value of n_component for 95% variance : {pca.n_components_}')

pca = PCA(n_components=0.90).fit(X)
print(f'Best value of n_component for 90% variance : {pca.n_components_}')
```

```python
pca = PCA(n_components=11).fit(X)

# Get the names of the top 11 principal components
top_11_components = pca.components_

# Get the column names of the original DataFrame X
column_names = X.columns

# Create a dictionary to map component indices to their corresponding feature names
component_feature_map = {}
for i, component in enumerate(top_11_components):
    component_name = f'Component {i+1}'
    top_feature_indices = np.abs(component).argsort()[-1::-1][:3]  # Get the top 3 features for each component
    top_features = [column_names[idx] for idx in top_feature_indices]
    component_feature_map[component_name] = top_features

# Print the results
for component_name, top_features in component_feature_map.items():
    print(f"{component_name}: {', '.join(top_features)}")
```

```python
# Since it is a very high dimensional data, we need to apply dimensionality reduction technique to only use important features.
# It can be clearly interpreted from this curve that only 11 components (features) accounts for 99% of the variance in data. 
# In other words we can also say that 99% of the information present in data can be obtained by using merely those top 11 
# features instead of 95 features.
# So I will transform my data using PCA.
```

```python
X_pca = PCA(11).fit_transform(X)
```

# The below code is to balance the data

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
#transforming the data with SMOTE:

X_train, X_test, y_train, y_test = train_test_split(X_pca,y,test_size=0.25, random_state=42 )
print("Before Oversampling : ")
print(f"Bankrupt 1 = {sum(y_train==1)}")
print(f"Bankrupt 0 = {sum(y_train==0)}")
print("========="*6)

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("After Oversampling : ")
print(f"Bankrupt 1 = {sum(y_train_smote==1)}")
print(f"Bankrupt 0 = {sum(y_train_smote==0)}")


# SMOTE generates synthetic samples for the minority class to balance the class distribution.
# This applies SMOTE to the features (X) and target variable (y) and generates synthetic samples for the minority class. 
# It returns the resampled features (X_sm) and target variable (y_sm) with a balanced class distribution.
# which can lead to better performance of your model, especially for classification tasks.
```

```python
# Create a count plot
sns.countplot(x=y_train_smote)

# Add labels and title
plt.xlabel('Target Labels')
plt.ylabel('Count')
plt.title('Count of Target Labels after SMOTE Resampling')

# Show the plot
plt.show()
```

```python
y_train_smote.value_counts()   #balanced data 
```

# splitting the data

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_smote, y_train_smote, test_size = 0.2, random_state = 0)

# splits the resampled features (X_sm) and target variable (y_sm) into training and testing sets. 
# The test_size parameter specifies the proportion of the dataset to include in the test split (in this case, 20%), 
# and the random_state parameter ensures reproducibility by fixing the random seed.

# we can use X_train and y_train for training your machine learning model 
# and X_test and y_test for evaluating its performance on unseen data.
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

# SVM (SUPPORT VACTOR MACHINE)

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto',C=20, kernel='rbf', probability=True)

svc.fit(X_train,y_train)
svc_pred = svc.predict(X_test)
print("Train score: {}" .format(svc.score(X_train,y_train)))
print("Accuracy score: {}" .format(accuracy_score(y_test,svc_pred)))
print(classification_report(y_test, svc_pred))
sns.heatmap(confusion_matrix(y_test,svc_pred), annot = True, cmap='Blues',fmt = 'd')
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
rfc_pred = rfc.predict(X_test)
print("Train score: {}" .format(rfc.score(X_train,y_train)))
print("Accuracy score: {}" .format(accuracy_score(y_test,rfc_pred)))
print(classification_report(y_test, rfc_pred))
sns.heatmap(confusion_matrix(y_test,rfc_pred), annot = True, cmap='Blues',fmt = 'd')
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_test)
print("Train score: {}" .format(lr.score(X_train,y_train)))
print("Accuracy score: {}" .format(accuracy_score(y_test,lr_pred)))
print(classification_report(y_test, lr_pred))
sns.heatmap(confusion_matrix(y_test,lr_pred), annot = True, cmap='Blues',fmt = 'd')
```

# KNN (K-Nearest Neighbour)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4, algorithm='auto')
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)
print("Train score: {}" .format(knn.score(X_train,y_train)))
print("Accuracy score: {}" .format(accuracy_score(y_test,knn_pred)))
print(classification_report(y_test, knn_pred))
sns.heatmap(confusion_matrix(y_test,knn_pred), annot = True, cmap='Blues',fmt = 'd')
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import  GradientBoostingClassifier

# Gradient Boosting Classifier
gradient_boosting = GradientBoostingClassifier(n_estimators=100, random_state=42)
gradient_boosting.fit(X_train, y_train)
gradient_boosting_pred = gradient_boosting.predict(X_test)

print("\nGradient Boosting Classifier:")
print("Train score:", gradient_boosting.score(X_train, y_train))
print("Accuracy score:", accuracy_score(y_test, gradient_boosting_pred))
print(classification_report(y_test, gradient_boosting_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, gradient_boosting_pred)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted labels')
plt.ylabel('Actual labels')
plt.title('Confusion Matrix - Gradient Boosting Classifier')
plt.show()
```

# Comparison of Model Performance

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate ROC curve and AUC for each model
fpr_svc, tpr_svc, _ = roc_curve(y_test, svc_pred)
auc_svc = roc_auc_score(y_test, svc_pred)

fpr_rfc, tpr_rfc, _ = roc_curve(y_test, rfc_pred)
auc_rfc = roc_auc_score(y_test, rfc_pred)

fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_pred)
auc_lr = roc_auc_score(y_test, lr_pred)

fpr_knn, tpr_knn, _ = roc_curve(y_test, knn_pred)
auc_knn = roc_auc_score(y_test, knn_pred)

fpr_gb, tpr_gb, _ = roc_curve(y_test, gradient_boosting_pred)
auc_gb = roc_auc_score(y_test, gradient_boosting_pred)

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_svc, tpr_svc, label=f'SVM (AUC = {auc_svc:.2f})', color='blue')
plt.plot(fpr_rfc, tpr_rfc, label=f'Random Forest (AUC = {auc_rfc:.2f})', color='green')
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {auc_lr:.2f})', color='orange')
plt.plot(fpr_knn, tpr_knn, label=f'KNN (AUC = {auc_knn:.2f})', color='red')
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC = {auc_gb:.2f})', color='cyan')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Plot random guessing line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Define the model names and their accuracy scores
models = ['SVM', 'Random Forest', 'Logistic Regression', 'KNN', 'Gradient Boosting']
accuracy_scores = [svc.score(X_test, y_test), rfc.score(X_test, y_test), lr.score(X_test, y_test), 
                   knn.score(X_test, y_test), gradient_boosting.score(X_test, y_test)]

# Plot the bar plot
plt.figure(figsize=(10, 5))
plt.bar(models, accuracy_scores, color=['blue', 'green', 'orange', 'red', 'cyan'])
plt.xlabel('Model')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Score Comparison of Different Models')
plt.ylim(0.8, 1.0)  # Set the y-axis limits to focus on the differences
plt.xticks(rotation=45)
plt.show()
```
